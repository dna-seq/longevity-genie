```python
import click
import polars as pl
from click.core import Context

from genie.config import Locations
from pathlib import Path
```

```python
base = Path("..")
print(base.absolute())
```

```python
locations = Locations(base)
locations
```

```python
clinvar = pl.read_csv(locations.clinvar_text, sep="\t")
clinvar
```

```python
report = pl.read_csv(locations.report_anton, sep="\t")
report
```

```python
clinical_report = report.with_columns([pl.col("snp").alias("rsid")]).join(clinvar,on="rsid").sort(pl.col("category_name"))
clinical_report
```

```python
cat = clinical_report.select(["rsid", "category_name"])
cat
```
